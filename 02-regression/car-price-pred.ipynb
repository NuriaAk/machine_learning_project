```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sn
```

```python
df = pd.read_csv('data.csv')
df.head()
```

# Cleaning

consistency

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
# Find all str cols
str_cols = list(df.dtypes[df.dtypes == 'object'].index)
```

```python
for col in str_cols:
    df[col] = df[col].str.lower().str.replace(' ','_')

df.head()
```
